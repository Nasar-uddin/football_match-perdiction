```python
matches = df.shape[0]
avg_home_scored = df.FTHG.sum() / matches
avg_away_scored = df.FTAG.sum() / matches
avg_home_conceded = avg_away_scored
avg_away_conceded = avg_home_scored

print(f"Total number of matches: {matches}")
print(f"Average number of goals at home: {avg_home_scored}")
print(f"Average number of goals away: {avg_away_scored}")
print(f"Average number of goals conceded at home: {avg_home_conceded}")
print(f"Average number of goals conceded away: {avg_away_conceded}")
```

```python
table = pd.DataFrame(columns=('Team','HAS','AAS','HDS','ADS','HGS','AGS','HGC','AGC'))
```

```python
home= df.groupby('HomeTeam') 
away= df.groupby('AwayTeam')
```

```python
table.Team = home.HomeTeam.first().values

table.HGS = home.FTHG.sum().values
table.HGC = home.FTAG.sum().values



table.AGS =away.FTAG.sum().values
table.AGC = away.FTHG.sum().values


#19 Home matches for each team each season and 16 seasons therefore 304 home matches and 304 away matches
table.HAS = (table.HGS / home.HomeTeam.count().values) / avg_home_scored
table.AAS = (table.AGS / away.AwayTeam.count().values) / avg_away_scored
table.HDS = (table.HGC / home.HomeTeam.count().values) / avg_home_conceded
table.ADS = (table.AGC / away.AwayTeam.count().values) / avg_away_conceded
```

```python
team_group=[]
h2h_hw={}
h2h_aw={}
drawn={}

for index,row in df.iterrows():
    x=row.HomeTeam
    y=row.AwayTeam
    team_pair=x+","+y
    
    def checking():
        a,b=team_pair.split(',')
        team_pair2=b+","+a
        if team_pair in team_group:
            return team_pair
        elif team_pair2 in team_group:
            return  team_pair2
        return -1
       
  
        
    if checking() in team_group:
        
        home,away=checking().split(',')
        
        if row.FTR == 'H':
            if row.HomeTeam == home:
                h2h_hw[checking()] = h2h_hw[checking()]+1
            else:
                h2h_aw[checking()] = h2h_aw[checking()]+1
                
        if row.FTR == 'A':
            if row.AwayTeam == away:
                h2h_aw[checking()] = h2h_aw[checking()]+1
            else:
                h2h_hw[checking()] = h2h_hw[checking()]+1
                
        if row.FTR =='D':
            drawn[checking()]=drawn[checking()]+1
    else:
        
        team_group.append(team_pair)
        home,away=team_pair.split(',')
        if row.FTR == 'H':
            if row.HomeTeam == home:
                h2h_hw[checking()]=1
                h2h_aw[checking()]=0
            else:
                h2h_aw[checking()]=1
                h2h_hw[checking()]=0
            drawn[checking()]=0
            
        if row.FTR == 'A':
            if row.AwayTeam == away:
                h2h_aw[checking()]=1
                h2h_hw[checking()]=0
            else:
                h2h_hw[checking()]=1
                h2h_aw[checking()]=0
            drawn[checking()]=0
            
        if row.FTR =='D':
            h2h_hw[checking()]=0
            h2h_aw[checking()]=0
            drawn[checking()]=1
        
        

# print(h2h_hw)

Team=[]
H2H_HW=[]
H2H_AW=[]
Drawn=[]
for i,j in h2h_hw.items():
        a,b=i.split(',')
        k=f"{a},{b}"
        Team.append(k)
        H2H_HW.append(j)
    


for i,j in h2h_aw.items():
        H2H_AW.append(j)
for i,j in drawn.items():
        Drawn.append(j)


def c(Team,H2H_HW,H2H_AW,Drawn):
    return [list(c) for c in zip(Team,H2H_HW,H2H_AW,Drawn)]

features = c(Team,H2H_HW,H2H_AW,Drawn)
```
